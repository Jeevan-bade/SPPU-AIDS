```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")
```

```python
titanic.isnull().sum()
```

```python
# Overall survival cout
plt.figure(figsize = (6, 4))
sns.countplot(data = titanic, x = "survived")
plt.title("Pattern : Overall Survival count (0 -> Died 1 -> survived)")
plt.xlabel("Survived")
plt.ylabel("Passenger Count")
plt.show()
```

```python
# Survival by Gender
plt.figure(figsize = (6, 4))
sns.countplot(data = titanic, x = "sex", hue = "survived")
plt.title("Pattern : Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Passengr Count")
plt.show()
```

```python
# Survival by Passenger class
plt.figure(figsize = (6, 4))
sns.countplot(data = titanic, x = "pclass", hue = "survived")
plt.title("Pattern : Survival by Passenger class")
plt.xlabel("Passenger class (1 -> 1st class)")
plt.ylabel("Passenger count")
plt.legend(title = "survived")
plt.show()
```

```python
# Survival by age distribution
plt.figure(figsize = (8, 4))
sns.histplot(data = titanic, x = "age", hue = "survived", bins = 30, kde = True)
plt.title("Pattern : Survival by age distribution")
plt.xlabel("Age")
plt.ylabel("Passenger count")
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.heatmap(titanic.isnull(), cbar = False, cmap = "viridis", yticklabels = "False")
plt.title("Heatmap missing values")
plt.show()
```

```python
plt.figure(figsize = (8, 6))
sns.heatmap(titanic.corr(numeric_only = True), cmap = "coolwarm", linewidths = 0.5)
plt.title("Correlation Heatmap of Numeric features")
plt.show()
```

```python
plt.figure(figsize = (8, 4))
sns.histplot(data = titanic, x = "fare" , bins = 40, kde = True, color = "teal")
plt.title("Pattern : Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Passenger count")
plt.show()
```
