```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# 1. List features and types
print("Features and types:")
for feature in iris.feature_names:
    print(f"{feature}: numeric (continuous)")

# 2. Histograms for each feature
iris_df.hist(figsize=(12, 8))
plt.suptitle("Histograms of Iris Features", y=1.02)
plt.show()

# 3. Box plots for each feature
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris_df.drop('species', axis=1))
plt.title("Box Plots of Iris Features")
plt.ylabel("Measurement (cm)")
plt.show()

# 4. Compare distributions and identify outliers
for feature in iris.feature_names:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='species', y=feature, data=iris_df)
    plt.title(f"Distribution of {feature} by Species")
    plt.show()
```
